```python
from sympy.interactive import printing
import numpy as np
printing.init_printing(use_latex=True)
import sympy as sym
%matplotlib inline
from matplotlib import pyplot as plt
from sympy import *

x, y = sym.symbols("x y")
```

```python
#Zad1
exp = (-x**3 + 3*x**2 + 10*x -24)
exp
eq = Eq(exp,0)
print("Punkty Przeciecia")
print(solve(eq,x))
xArr = np.arange(-5, 5, 0.1)
y = [exp.evalf(subs={x: i}) for i in xArr]
v = np.linspace(-5,5,100)
plt.plot(v,y)
plt.show()
```

```python
#Zad2
row1 = Eq(x**2 + 3*y,10)
row2 = Eq(4*x - y**2,-2)
solved = solve([row1,row2],[x,y])
solved
```

```python
#Zad3
print("Uklad ma {} rozwiazania".format(len(solved)))
```
